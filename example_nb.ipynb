## Trained model in Action

```python
from envs.rubiks_cube_env import RubiksCubeEnv
```

Change the parameter scramble_moves to load the rubiks cube with desired no of scramble moves away from solved state

```python
myenv = RubiksCubeEnv(scramble_moves=3)
myenv.reset()
myenv.render()
```

Load the desired trained model from the results/models folder

```python
from stable_baselines3 import PPO
```

```python
model = PPO.load("results/models/PPO_3.zip", env=myenv)
```

Use the trained model for solving the Rubik's cube

```python
done = False
i = 0
while not done:
    action, _states = model.predict(myenv.state)
    obs, reward, done, truncated, info = myenv.step(action)
    print(f"Action: {action}, Reward: {reward}, Done: {done}")
    myenv.render()
    i += 1
    if done:
        print("Episode finished")
        print(f"Total steps: {i}")
        print(f"Total reward: {reward}")
        break
myenv.close()
```
